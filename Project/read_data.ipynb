```python
import matplotlib.pyplot as plt

# Adjust the orientation of the ellipse for age 10 to match the others
fig, ax = plt.subplots()

# Set axis labels
ax.set_xlabel('Exercise')
ax.set_ylabel('Cholesterol')

# Adjusted ranges to prevent ellipses from going outside the frame
exercise_range = [20, 30, 40, 50, 60]
cholesterol_range = [15, 25, 35, 45, 55]
heights = [20,35,40,45,50]
widths = [20,20,15,20,20]

# Create overlapping ellipses with consistent orientation
for i, age in enumerate([10, 20, 30, 40, 50]):
    
    ellipse = plt.matplotlib.patches.Ellipse((exercise_range[i], cholesterol_range[i]), width=widths[i], height=heights[i]*1.2, angle=40, alpha=0.3, color='blue')
    ax.add_patch(ellipse)
    ax.text(exercise_range[i] + 2, cholesterol_range[i] + 2, str(age), fontsize=12)

# Set limits and title, ensuring all ellipses are visible
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_title('Cholesterol vs Exercise')

# Display the plot
plt.show()
```
